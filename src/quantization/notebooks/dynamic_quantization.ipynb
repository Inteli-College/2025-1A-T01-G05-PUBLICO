```python
# Libraries

# Built-in
from pathlib import Path

# Third-party
from onnxruntime.quantization import quantize_dynamic, QuantType
from ultralytics import YOLO
```

```python
model_fp32 = Path('models/best_fp32_preprocessed.onnx')
static_model_quant = Path('models/dynamic_int8_quant.onnx')
validation_data = Path('tcc-1/data.yaml')
```

```python
# Quantization process
quantized_model = quantize_dynamic(
                model_fp32,
                static_model_quant,
                weight_type=QuantType.QUInt8
            )
```

```python
model = YOLO(static_model_quant, task="detect")
metrics = model.val(data=validation_data, split='val')
print(metrics)
```
